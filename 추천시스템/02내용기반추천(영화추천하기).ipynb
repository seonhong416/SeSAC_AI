```python
# CountVectorizer 적용하기 위해 공백문자로 word 단위로 구분되는 문자열로 변환

df['genre_literal'] = df['genres'].apply(lambda x : (' ').join(x))
df['genre_literal'][:1]
```

```python
# CountVectorizer로 학습

count_vec = CountVectorizer(min_df = 0, ngram_range = (1,2))
# min_df : 단어장에 들어갈 최소 빈도(초기화), ngram_range = (1, 2) : 1 <= n <= 2 (onegram or bigram)
```

```python
count_vec.fit_transform(df['genre_literal'])
# 4803*276 (4803 : 영화 갯수) >> sparse matrix(행렬)
```

```python
genre_mat = count_vec.fit_transform(df['genre_literal'])
print(genre_mat.shape)
```

```python
print(genre_mat) # onegram, bigram (0번째 onegram5개 bigram4개 > 총 9개)
```

```python
# 코사인 유사도 이용, 영화별 유사도 계산

from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
genre_sim
```

```python
# 실제 함수

def find_sim_movie_ver1(df, sorted_ind, title_name, top_n = 10) :
    # 인자로 입력된 df 데이터프레임에서 'title' 컬럼이 입력되니 title_name 값인 데이터 프레임 추출
    
    title_movie = df[df['title'] == title_name]
    
    # title_name을 가진 데이터프레임의 index객체를 ndarray로 반환하고
    # sorted_ind 인자로 입력된 genre_sim_sorted_ind 객체에서 유사도 순으로 top_n개의 index 추출
    
    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, : (top_n)]
    
    # 추출된 top_n index를 출력(2차원)
    # 데이터프레임에서 index로 사용하기 위해서는 1차원 array로 변경
    
    print(similar_indexes)
    
    # 2차원 >> 1차원
    similar_index = similar_indexes.reshape(-1)
    
    return df.iloc[similar_index]
```

```python
# 영화 God Father와 장르가 유사한 영화 10개 추천

similar_movies = find_sim_movie_ver1(df, genre_sim_sorted_ind, 'The Godfather', 10)
similar_movies
```

```python
# 평점기준으로 정렬(sort_values)해 보니, 평가횟수(vote_count)가 낮은 것들도 파함되어 있음
# 평가 횟수 반영한 평점(가중평점)이 필요함. 내림차순으로 상위 10개만 추출할 필요있음.
df[['title', 'genres', 'vote_average', 'vote_count']].sort_values('vote_average', ascending=False)[:10]
```

```python
# 가중평점(평점, 평가횟수) 반영한 영화 추천
# (v/(v+m))* R + (m/(v+m))*C
# 상위 60% 해당하는 vote_count 최소 투표 횟수인 m 으로 지정
# v : 영화별 평점을 투표한 횟수
# m : 평점을 부여하기 위한 최소 투표 횟수
# R : 개별 영화에 대한 평균 평점
# C : 전체 영화에 대한 평균 평점



C = df['vote_average'].mean()
# 전체 영화에 대한 평점 6.09

m = df['vote_count'].quantile(0.6)
# 370.199999
# >> 적어도 370번 이상의 평점을 준 영화의 평점을 사용
```

```python
# C : 전체 영화에 대한 평균 평점 : 약 6점
# m : 평점을 부여하기 위하나 최소 투표 회수 : 370회(상위 60% 수준)
print('C : ', round(C, 2), '\nm : ' , round(m, 2))
```

```python
# 가중 평점을 계산하는 함수
def weighted_vote_average(record) :
    v = record['vote_count']
    R = record['vote_average']
    
    return ((v/(v+m))*R + (m/(m+v))*C)
```

```python
# 기존 데이터에 가중평점 컬럼 추가

df['weighted_vote'] = df.apply(weighted_vote_average, axis = 1)
df['weighted_vote']
```

```python
# 가중평균 기준, 영화 순위 추출(내림차순) >> 평점이 높은 영화 먼저 출력

df.columns
```

```python
df[['title', 'genres', 'weighted_vote', 'vote_average', 'vote_count']].sort_values('weighted_vote', ascending=False)[:10]
```

```python
# 추천 ver2 먼저 장르 유사성 높은 영화 선정 후 가중평점 순으로 선정

def find_sim_movie_ver2(df, sorted_ind, title_name, top_n = 10) :
    title_movie = df[df['title'] == title_name]
    title_index = title_movie.index.values
    
    # top_n의 2배(유사성 높은 영화 선정)에 해당하는 장르 유사성이 높은 index 추출
    similar_indexes = sorted_ind[title_index, : (top_n)*2]
    similar_index = similar_indexes.reshape(-1)
    
    # 기준 영화 index는 제외(기존에 본 영화 제외하고 추천하기 위함, title_index 겹치지 않는 영화 추천)
    similar_index = similar_index[similar_index != title_index]
    
    # top_n의 2배(후보군)에서 weighted_vote가 높은 순으로 top_n만큼 추출
    return df.iloc[similar_index].sort_values('weighted_vote', ascending = False)[:top_n]

```

```python
# 영화 Godfather 에 대해 장르 유사성, 가중평점을 반영한 영화 10편 추천

similar_movies = find_sim_movie_ver2(df, genre_sim_sorted_ind, 'The Godfather', 10)
similar_movies[['title', 'genres', 'weighted_vote', 'vote_average', 'vote_count']]
```
