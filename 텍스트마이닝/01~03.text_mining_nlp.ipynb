```python
import nltk
nltk.download('stopwords') # 불용어
```

```python
from nltk.tokenize import RegexpTokenizer
RegexpTokenizer(r'\w+')
tokenizer = RegexpTokenizer(r'\w+')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords

stop_words = nltk.corpus.stopwords.words('english')
stop_words
```

```python
stop_words.extend([])
stop_words[:10]
```

```python
from nltk import wordnet
nltk.download('wordnet')
```

```python
nltk.download('omw-1.4')
```

```python
# 전처리

preprocessed_texts = []

for text in texts.values :
    tokenized_text = tokenizer.tokenize(text.lower()) 
    # print(tokenized_text)
    
    lemmatized_text = [lemmatizer.lemmatize(token) for token in tokenized_text]
    # print(lemmatized_text)
    
    stopped_text = [token for token in lemmatized_text if token not in stop_words]
    # print(stopped_text)
    
    preprocessed_texts.append(stopped_text)
    
    
    
```

```python
len(preprocessed_texts)
```

```python
from collections import Counter

tokens = []

for text in preprocessed_texts : 
    tokens.extend(text) 
    
counted_tokens = Counter(tokens)
# print(counted_tokens)
```

```python
top_20 = counted_tokens.most_common(20)
top_20
```

```python
from konlpy.tag import Kkma, Komoran, Okt

okt = Okt()
kkm = Kkma()
kom = Komoran()

from eunjeon import Mecab
mecab = Mecab()
```

```python
# 전처리

# 1. 불용어 제거(stopwords)

stop_words = '은 이 것 등 더 를 좀 즉 인 옹 때 만 원 이때 개 일 기 시 럭 갤 성 삼 스 폰 트 드 기 이 리 폴 사 전 마 자 플 블 가'
stop_words = stop_words.split()
stop_words[:10]
```

```python
# 2. 형태소 별로 모으기

# 명사모으기(okt 활용)

nouns = []

for p in post :
    for noun in okt.nouns(p) :
        if noun not in stop_words :
            nouns.append(noun)

nouns
```

```python
# 명사모으기 (mecab활용)

nouns_m = []

for p in post :
    for noun in mecab.nouns(p) :
        if noun not in stop_words :
            nouns_m.append(noun)

nouns_m
```

```python
# 단어 모으기(pos tagging 포함) (okt활용)

words = []

for p in post :
    words.extend(okt.pos(p))
words[:10]
```

```python
# 단어모으기(mecab활용)
words_m = []

for p in post :
    words_m.extend(mecab.pos(p))
words_m[:10]
```

```python
# 동사 모으기(okt 활용)

verbs = []

for word in words :
    if word[1] == 'Verb' :
        verbs.append(word[0])

verbs
```

```python
# 동사모으기(mecab 활용)
verbs_m = []
for word in words_m :
    if word[1] == 'VV' :
        verbs_m.append(word[0]+'다')

verbs_m[:10]
```

```python
# 형용사모으기(okt활용)

adjective = []

for word in words :
    if word[1] == 'Adjective' :
        adjective.append(word[0])
        
adjective[:10]
```

```python
# 형용사 모으기(mecab활용)
adjective_m = []

for word in words_m :
    if word[1] == 'VA' :
        adjective_m.append(word[0])
    
adjective_m[:10]
```

```python
from collections import Counter
counted_nouns = Counter(nouns)

top_20 = counted_nouns.most_common(20)
top_20
```
