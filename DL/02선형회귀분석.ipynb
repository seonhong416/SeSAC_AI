```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2,4,6,8])
y = np.array([81, 93, 91, 97])
```

```python
plt.scatter(x, y)
plt.show()
```

```python
# 기울기 w, 절편 b 초기화
w = 0 
b = 0

# 학습률 (임의로 0.03)
lr = 0.03

# 반복 횟수
epochs = 2001
```

```python
# 경사하강법

n = len(x)

# 경사하강법 시작

for i in range(epochs) :
    y_pred = w * x + b
    error = y - y_pred
    
    w_diff = (2/n)*sum(-x * (error)) # 오차함수를 w로 편미분한 값
    b_diff = (2/n)*sum(-(error))     # 오차함수를 b로 편미분한 값
    
    w = w - lr*w_diff # 학습률 곱해서 기존의 w를 업데이트
    b = b - lr*b_diff # 학습률 곱해서 기존의 b를 업데이트
    
    if i % 100 == 0 :
        print(f'epoch : {i} \t기울기 : {w} \t절편 : {b}')
```

```python
y_pred = w*x + b

plt.scatter(x, y)
plt.plot(x, y_pred, color = 'r')
plt.show()
```

```python
# 다중 선형 회귀
x1 = np.array([2,4,6,8])
x2 = np.array([0,4,2,3])
y = np.array([81, 93, 91, 97])
```

```python
# 데이터 분포 확인

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax = ax.scatter3D(x1, x2, y)
plt.show()
```

```python
# 변수 설정

w1 = 0
w2 = 0
b = 0
lr = 0.01

epochs = 4001
```

```python
# 경사하강법

n = len(x)

for i in range(epochs) :
    y_pred = w1*x1 + w2*x2 + b
    error = y - y_pred
    
    # 편미분
    w1_diff = (2/n)*sum(-x1 * error)
    w2_diff = (2/n)*sum(-x2 * error)
    b_diff = (2/n)*sum(-error)
    
    w1 = w1 - lr * w1_diff
    w2 = w2 - lr * w2_diff
    b = b - lr * b_diff
    
    if i% 100 == 0 :
        print(f'epoch : {i} \t기울기 w1 : {w1} \t기울기 w2 : {w2} \t절편 : {b}')
```

```python
# 텐서플로우 활용, 선형 회귀 모델

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([2,4,6,8])
y = np.array([81, 93, 91, 97])
```

```python
model = Sequential()

# 출력 값, 입력 변수, 분석방법(활성화 함수) 설정
model.add(Dense(1, input_dim = 1, activation = 'linear'))

# 오차 수정을 위해 경사하강법(sgd) 사용
# 오차의 정도를 판단하기 위한 평가방법 >> mse(평균제곱오차) 사용
model.compile(optimizer = 'sgd', loss = 'mse')

# 오차를 최소화하는 과정을 2000번 반복

model.fit(x, y, epochs = 2001, )
```
