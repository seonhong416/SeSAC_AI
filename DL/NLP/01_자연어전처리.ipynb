```python
# 불용어_stopwords (형태소 분석기 돌리고 난 뒤에 정함)
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']

mecab = Mecab()
mecab.morphs('와 이런 것도 영화라고??!?')
```

```python
x_train = []

for sentence in tqdm(train['document']) :
    tokenized_sentence = mecab.morphs(sentence) # 토큰화
    stopwords_removed_sentence = [word for word in tokenized_sentence if not word in stopwords] # 불용어 제거
    x_train.append(stopwords_removed_sentence)
```

```python
x_test = []

for sentence in tqdm(test['document']) :
    tokenized_sentence = mecab.morphs(sentence) # 토큰화
    stopwords_removed_sentence = [word for word in tokenized_sentence if not word in stopwords] # 불용어 제거
    x_test.append(stopwords_removed_sentence)
```

```python
word_list = []

for sent in x_train :
    for word in sent :
        word_list.append(word)
        

word_counts = Counter(word_list)

print('총 단어수 :', len(word_counts))
```

```python
print(word_counts)
```

```python
print('훈련 데이터에서 단어 영화의 등장 횟수 :', word_counts['영화'])
print('훈련 데이터에서 단어 송강호의 등장 횟수 :', word_counts['송강호'])
print('훈련 데이터에서 단어 열외의 등장 횟수 :', word_counts['열외'])
```

```python
vocab = sorted(word_counts, key=word_counts.get, reverse=True)
print('등장 빈도수 상위 10개 단어')
print(vocab[:10])
```

```python
threshold = 3 # 임계치(기준)
# 등장빈도가 threshold 값인 3회 미만, 즉 2회이하인 단어들이 단어 집합에서 몇 % 차지하는 지를 보고
# 정수 인코딩 과정에서 포함할지 배제할 지 정하고자 함
# 등장 빈도가 threshold 미만인 단어들을 제외한 단어를 가지고 정수 인코딩 함

total_cnt = len(word_counts) # 단어 수
rare_cnt = 0
# 등장 빈도수가 threshold 보다 작은 단어의 개수를 카운트 하기 위해 초기화
total_freq = 0
# 훈련 데이터의 전체 단어 빈도수의 총 합 구하기 위해 초기화
rare_freq = 0
# 등장 빈도수가 threshold 보다 작은 단어의 등장 빈도수의 총 합 초기화

# 단어와 빈도수의 쌍(pair)을 key와 value로 받는다

for key, value in word_counts.items() :
    total_freq = total_freq + value
    
    # 단어의 등장 빈도수가 threshold보다 작으면
    if value < threshold :
        rare_cnt += 1
        rare_freq += value

print('단어 집합(vocabulary) 크기 :', total_cnt)
print('등장 빈도가 %s번 이하인 희소 단어의 수 : %s'%(threshold-1, rare_cnt))
print(f'단어 집합에서 희소 단어의 비율 : {(rare_cnt/total_cnt)*100:.2f}%')
print(f'전체 등장 빈도에서 최소 단어 등장 빈도 비율 : {(rare_freq/total_freq)*100:.2f}%')
        
```

```python
# 전체 단어 개수 중에서 빈도수 2이하인 단어는 제거( 등장빈도수로 정렬된 상태임)

vocab_size = total_cnt - rare_cnt
print(vocab[:vocab_size])
```

```python
vocab = vocab[:vocab_size]
print('단어의 집합 크기 :', len(vocab))
```

```python
word_to_index = {}
word_to_index['<PAD>'] = 0
word_to_index['<UNK>'] = 1
```

```python
def texts_to_sequences(tokenized_x_data, word_to_index) :
    encoded_x_data = []
    for sent in tokenized_x_data :
        index_sequences = []
        for word in sent :
            try :
                index_sequences.append(word_to_index[word])
            
            except KeyError :
                index_sequences.append(word_to_index['<UNK>'])
        
        encoded_x_data.append(index_sequences)
    return encoded_x_data
```

```python
encoded_x_train = texts_to_sequences(x_train, word_to_index)

encoded_x_train[:10]
```

```python
encoded_x_test = texts_to_sequences(x_test, word_to_index)
encoded_x_test[:10]
```
