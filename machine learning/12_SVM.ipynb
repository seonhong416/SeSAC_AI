```python
# 서포트 벡터 머신
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import svm, datasets
```

```python
x = [[0,0], [1,1]]
y = [0, 1]

clf = svm.SVC() # SVC : support vector classifier
clf.fit(x, y)
```

```python
# 데이터 시각화

plt.scatter([0],[0], c = 'red')
plt.scatter([1],[1], c = 'blue')

plt.xlim(-2, 3)
plt.ylim(-2,3)
plt.xlabel('x[0]')
plt.ylabel('x[1]')
plt.grid()
plt.show()
```

```python
# test data로 예측 >> 첫번째 클래스로 예측

clf.predict([[-1,-1]])
```

```python
clf.predict([[2,2]])
```

```python
### support vector 확인
# >> support vector point 알려줌
clf.support_vectors_
```

```python
# support vectors의 data index
# 데이터 중에서 몇번째 데이터가 서포트 벡터인지 확인
# 여기서는 0번, 1번 데이터가 서포트 벡터
clf.support_
```

```python
# 클래스 별 support vectors의 개수
# 0번째 클래스에 1개, 1번째 클래스에 1개
clf.n_support_
```

```python
# SVC 분류 예제 (iris 데이터)

iris = datasets.load_iris()
iris['data'][:10]
```

```python
# 0번, 1번 꽃의 꽃받침(sepal)의 길이(length)와 너비(width)만 저장

x_iris = iris.data[:100, :2]
y_iris = iris.target[:100]
```

```python
x_iris
```

```python
y_iris
```

```python
# SVM의 정규화
# sklearn의 C는 SVM 수식의 C와 반비례 관계
# 즉, C가 클수록 더 적은 수의 support vector 허용
# cf) SVM 수식의 C는 클수록 더 많은 support vector 가짐

C = 100.0

# C = 10
```

```python
# 훈련용 데이터로 적합

clf = svm.SVC(kernel = 'linear', C = C).fit(x_iris, y_iris)
```

```python
# 어느 것이 support vector인지 확인

clf.support_vectors_
```

```python
# 이 support vector가 몇 번째 데이터 인지 확인
clf.support_
```

```python
x_iris[clf.support_[0]]
# iris data(훈련용 데이터) 중 36번째 해당하는 데이터 가져와 라는 의미
```

```python
# 클래스별 support vectors의 개수

clf.n_support_
```

```python
# SVC 시각화

# 그림 크기 설정
plt.figure(figsize=(8,6))

# 분류의 경계선과 배경 만들기
h = 0.2

x_min, x_max = x_iris[:, 0].min() - 1, x_iris[:, 0].max() + 1
y_min, y_max = x_iris[:, 1].min() - 1, x_iris[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

# z(예측 결과물) >> color plot
z = z.reshape(xx.shape)
plt.contourf(xx, yy, z, cmap = plt.cm.coolwarm, alpha = 0.8)

# 훈련용 데이터(training data) >> 배경 위에 표시(꽃받침의 길이와 너비)
plt.scatter(x_iris[:,0], x_iris[:, 1], c = y_iris, cmap = plt.cm.coolwarm)

# support vector를 배경 위에 표시
for sv_idx in clf.support_ :
    plt.scatter(x_iris[sv_idx][0], x_iris[sv_idx][1], c = 'white', marker ='s')
    
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks()
plt.yticks()
plt.title('SVC with linear kernel')
```

```python
# SVM kernel 비교

C = 1.0

# SVC
svc = svm.SVC(kernel = 'linear', C = C).fit(x_iris, y_iris)

# SVM(radial kernel)
rbf_svc = svm.SVC(kernel = 'rbf', gamma = 1.5, C = C).fit(x_iris, y_iris)

# SVM(polinomial kernel)
poly_svc = svm.SVC(kernel = 'poly', degree = 3, C = C).fit(x_iris, y_iris)
```

```python
# SVC 시각화

# 그림 크기 설정
plt.figure(figsize=(12, 3))

# 분류 경계선 배경 만들기
h = 0.2

x_min, x_max = x_iris[:, 0].min() - 1, x_iris[:, 0].max() + 1
y_min, y_max = x_iris[:, 1].min() - 1, x_iris[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# 각 그림의 제목(title) 포현

titles = ['SVC with linear kernel', 'SVC with RBF kernel', 'SVC with polynomial(degree = 3) kernel']

for i, clf in enumerate((svc, rbf_svc, poly_svc)) :
    plt.subplot(1,3, i+1)
    plt.subplots_adjust(wspace=0.4, hspace=0.4) # 그림과 그림사이를 띄워주는 역할
    
    z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    
    z = z.reshape(xx.shape)
    plt.contourf(xx, yy, z, cmap=plt.cm.coolwarm, alpha = 0.8)
    
    # 훈련용 데이터(training data) >> 배경 위에 표시(꽃받침의 길이와 너비)
    plt.scatter(x_iris[:,0], x_iris[:, 1], c = y_iris, cmap = plt.cm.coolwarm)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title(titles[i])
    plt.xticks()
    plt.yticks()

plt.show()
        
    
    
```
