```python
import numpy as np
import pandas as pd

raw = pd.read_csv('./scores.csv')
raw
```

```python
raw.index
```

```python
df = raw.copy()
df.index = 'i'+df.index.astype('str')
df.head()
```

인덱스로 행 데이터 추출하기

- df.loc[idx]
    - 시리즈 형태로 추출됨. 하나의 인덱스만 사용 가능
- df.loc[인덱스 리스트]
    - 데이터 프레임 형태로 추출됨. 한 개 이상의 인덱스 사용 가능

Series형태로 추출

```python
df.loc['i3']
```

```python
s1 = df.loc['i3']
print(type(s1))
print(s1.index)
```

```python
print(s1.values)
```

인덱스로 행, 열 추출하기



```python
# loc[idx,column]

df.head(3)
```

```python
df.loc['i1','kor']
```

특정 인덱스의 여러 컬럼 데이터 추출

```python
df.loc['i1',['name','kor']]
```

여러 인덱스의 특정 컬럼 데이터를 추출

```python
df.loc[['i1','i2','i3'],'name']
```

여러 인덱스의 여러 컬럼 가져오기

```python
# loc[idx list, colum list]

df.loc[['i0','i1','i2'],['name','math']]
```

모든 행에서 특정 열 가져오기
- loc[:,컬럼명]
- loc[:,컬럼리스트]

```python
df.loc[:, 'name']
```

```python
df.loc[:,['name', 'math']]
```

```python
df.loc[:,['name']]
```
