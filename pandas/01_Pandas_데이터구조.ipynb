데이터 프레임 만들기
- 마치, 엑셀의 sheet 동일한 개념(2차원 행열 구조를 가진 표 형태)

```python
# 리스트로 만들기
l1 = [
    ['drwill',20,'programmer'],
    ['zzuyu',20,'singer'],
    ['zazuha',23,'ballerina']
]

df = pd.DataFrame(l1, columns = ['name', 'age', 'job'], index = ['a','b','c'])
```

```python
df
```

```python
# 딕셔너리로 만들기

d1 = {'name' : ['drwill','zzuyu','kazuha'], 'age' : [20, 20, 25], 'job' : ['programmer', 'singer', 'ballerina']}

df = pd.DataFrame(d1, index = ['a', 'b', 'c'])
df
```
