```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = ['a','b','c','d','e']
y = [1,3,2,10,7]
```

```python
# 선(line) 그래프
plt.plot(x,y)
plt.show()
```

```python
# 막대그래프

plt.bar(x,y)
plt.show()
```

```python
#가로 막대 그래프

plt.barh(x, y)
plt.show()
```

```python
# 산점도

plt.scatter(x,y)
plt.show()
```

세 가지 그래프를 한번에... + 옵션추가

```python
# 한글폰트 깨짐 방지

from matplotlib import font_manager, rc

font_path = '../visualization/malgun.ttf'
font_name = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font_name)

# 마이너스 부호 깨짐 방지
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.scatter(x, y, label = 'scatter') # 산점도
plt.bar(x, y, label = 'bar') # 바그래프
plt.plot(x, y, label = 'line') # 선그래프
plt.title('3가지 그래프 콤보 차트', size = 15)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
