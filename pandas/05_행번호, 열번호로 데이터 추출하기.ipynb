```python
import numpy as np
import pandas as pd

raw = pd.read_csv('./scores.csv')
raw.head()
```

```python
df = raw.copy()
```

```python
df.index = 'i'+df.index.astype('str')
df.head()
```

행번호로 행 데이터 추출하기(iloc)

- df.iloc[행번호]
- df.iloc[행번호 리스트]
- df.iloc[행번호 슬라이싱]
- 음수(-) 사용시, 뒤에서부터

```python
# Series 형태로 추출하기
df.iloc[0]
```

```python
# 데이터 프레임 형태로 추출

df.iloc[[0,2,3]]
```

```python
df.iloc[[0]] 
```

행번호 행, 열 추출하기

```python
# iloc[행번호, 열번호]
df.iloc[0,0]
```

```python
# iloc[행번호, 열번호리스트]
df.iloc[0,[1,2]]
```

```python
# iloc[행번호리스트, 열번호]
df.iloc[[1,3,4],1]
```

```python
# iloc[행번호리스트, 열번호리스트]
df.iloc[[1,3,4],[0,2]]
```

```python
#iloc[행번호 슬라이싱, 열번호 슬라이싱]

# 0,1번째 행, 0,1번째 열 슬라이싱
df.iloc[0:2,0:2]
```

```python
# 1,3,5번째 행, 0,2번째 열 슬라이싱
df.iloc[1:6:2, :3:2]
```

```python
df.iloc[[-1]]
```

```python
# 마지막 행 1,3 열 슬라이싱
df.iloc[-1,1:4:2]
```

```python
# 모든 행, 1열

df.iloc[:, [1]]
```

```python
# 모든 행, 1,2 열

df.iloc[:,1:3]
```

```python
 df.iloc[:,[1,2]]
```
