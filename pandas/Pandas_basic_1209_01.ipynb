```python
import numpy as np
import pandas as pd
```

```python
print('DataFrame')

df = pd.DataFrame({
      'col1' : [2,1,9,8,7,4],
      'col2' : ['a','a','b',np.nan,'d','c'],
      'col3' : [0,1,9,4,2,3]
  })

df
```

```python
# 정렬

# 1. col1을 기준, 오름차순 정렬하기

df.sort_values('col1',ascending = True) # default : 오름차순
```

```python
df.sort_values('col1',ascending = False) # 내림차순
```

```python
df.sort_values(['col2','col1'],ascending = [True,False])
# col2 오름차순 먼저하고 col1 내림차순
```

집계함수

```python
data = {
    'korea' : [50,60,70],
    'math' : [10, np.nan, 40]
}
data
```

```python
pd.DataFrame(data, index = ['a','b','c'])
```

```python
df = pd.DataFrame(data, index = ['a','b','c'])
df
```

```python
# 각 컬럼별 데이터 개수

df.count() # axis = 0 : default
```

```python
# 각 행별 데이터 개수

df.count(axis = 1)
```

```python
# 각 컬럼별 최대값

df.max()
```

```python
# 각 컬럼별 최솟값

df.min()
```

```python
# 각 컬럼별 합계

df.sum()
```

```python
# NAN값 컬럼의 최솟값으로 대체

df['math'].fillna(df['math'].min())
```

```python
df_copy = df.copy()
```

```python
df_copy['math'] = df['math'].fillna(df['math'].min())
df_copy
```

```python
# NAN값 > 컬럼의 평균값으로 대체
df_copy2 = df.copy()

df_copy2['math'].fillna(df_copy2['math'].mean(), inplace = True)
df_copy2
```

그룹으로 묶기

```python
df = pd.DataFrame({
      'key' : ['a','b','c','a','b','c'],
      'data1' :[1,2,3,1,2,3],
      'data2' :[4,4,6,0,6,1]
  })
```

```python
print('DataFrame\n')
df
```

```python
# groupby

df.groupby('key').sum()
```

```python
# key, data1 기준, 합계 구하기
df.groupby(['key','data1']).sum()
```

```python
df.groupby(['key','data2']).sum()
```

```python
# aggregate 함수 이용, 요양통계량 산출

# 'key' 컬럼을 묶은 후, data1, data2, 각각의 최솟값, 중앙값, 최대값 출력

df.groupby('key').aggregate([min, np.median, max])
```

```python
# 데이터프레임 'key' 컬럼으로 묶고, data1의 최소값, data2는 합계 출력하세요

df.groupby('key').aggregate({'data1' : min, 'data2' : sum})
```
