```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup

import numpy as np
import pandas as pd
import time
```

```python

# 크롬 드라이버 실행
driver = webdriver.Chrome('chromedriver.exe')

# url 접속하기
url = 'https://comic.naver.com/webtoon/weekday'
driver.get(url)
```

```python
weekday = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

for i in range(2, 9) :
    web_list = []
    title_l = []
    author_l = []
    score_l = []
    genre_l = []
    age_l = []
    like_l = []
    latest_l = []
    latest_d = []
    driver.find_element(By.XPATH,f'//*[@id="content"]/ul/li[{i}]/a').click()
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    web_list = soup.select('ul.img_list > li > dl > dt > a')
    time.sleep(2)
    print(i)
    for j in range(len(web_list)) :

        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        score = soup.select('ul.img_list > li > dl > dd > div.rating_type > strong')[j].text
        
        driver.find_element(By.XPATH,f'//*[@id="content"]/div[4]/ul/li[{j+1}]/dl/dt/a').click()
        time.sleep(2)
        
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        
        title = soup.select('h2 > span.title')[0].text
        author = soup.select('h2 > span.wrt_nm')[0].text
        genre = soup.select('span.genre')[0].text
        age = soup.select('span.age')[0].text
        like = soup.select('em.u_cnt')[0].text
        latest_score = soup.select('div.rating_type > strong')[0].text
        latest_date = soup.select('td.num')[0].text
        driver.back()
        time.sleep(2)
        
        
        title_l.append(title)
        author_l.append(author)
        score_l.append(score)
        genre_l.append(genre)
        age_l.append(age)
        like_l.append(like)
        latest_l.append(latest_score)
        latest_d.append(latest_date)
        
        print(j)
    data = pd.DataFrame([title_l, author_l, score_l, genre_l, age_l, like_l,latest_l,latest_d])
    data = data.T
    data.columns = ['제목', '작가', '평점', '장르', '연령', '좋아요', '최근평점', '최근날짜']
    data.to_excel(f'{weekday[i-2]}_web.xlsx', index = False)
        
```
